```python
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

all_titles = []
all_company_names = []
all_locations = []
next_page_exists = True
page_number = 2
while next_page_exists:
    time.sleep(5)

    # Scroll down the page
    browser.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(5)  # wait for the page to load

    # Extract the job titles
    jobs = browser.find_elements("xpath","//div[contains(@class, 'full-width') and contains(@class, 'artdeco-entity-lockup__title') and contains(@class, 'ember-view')]")
    c = [job.text for job in jobs]
    print(c)
    all_titles.append(c)
    print(len(c))
    
    #extract company name
    company_name = browser.find_elements("xpath", "//div[contains(@class, 'artdeco-entity-lockup__subtitle') and contains(@class, 'ember-view')]")
    cn = []
    for i in company_name:
        cn.append(i.text)
    all_company_names.append(cn)
        
    #extract location
    location = browser.find_elements("xpath", "//ul[contains(@class, 'job-card-container__metadata-wrapper')]")
    loc = []
    for i in location:
        loc.append(i.text)
    all_locations.append(loc)
        
    try:
        # Try to find and click on the next page number
        next_button = browser.find_element("xpath",f"//button[./span[text()='{page_number}']]")
        next_button.click()
        page_number += 1
        if page_number == 40:
            break
    except NoSuchElementException:
        try:
            # If the next page number is not found, find and click on the ellipsis
            ellipsis = browser.find_element("xpath","//button[./span[text()='…']]")
            ellipsis.click()
        except NoSuchElementException:
            next_page_exists = False    
```

```python
titles = [title for titles in all_titles for title in titles]
company_names = [company_name for company_names in all_company_names for company_name in company_names]
locations = [location for locations in all_locations for location in locations]
print(len(titles))
print(len(company_names))
print(len(locations))
all_titles
```
